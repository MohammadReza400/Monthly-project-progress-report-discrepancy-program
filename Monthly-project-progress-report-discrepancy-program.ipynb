```python
from PIL import Image, ImageEnhance
import cv2
import numpy as np

def load_and_preprocess_image():
    # بارگذاری تصویر با OpenCV
    img = cv2.imread('C:/Users/parsian/Desktop/Python/tr.png')
    
    # تبدیل به تصویر خاکستری
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    
    # بهبود کنتراست تصویر
    gray = cv2.equalizeHist(gray)
    
    # حذف نویز با فیلتر گوسی
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    
    # ذخیره تصویر برای OCR
    cv2.imwrite("preprocessed_image.png", blurred)
    
    return blurred

# فراخوانی تابع و تست آن با یک تصویر نمونه
preprocessed_img = load_and_preprocess_image()
```

```python
import pandas as pd

def compare_tables(df_current, df_previous):
    # محاسبه تفاوت درصدها
    df_diff = df_current.copy()
    df_diff['Progress Difference'] = df_current['Progress (%)'] - df_previous['Progress (%)']
    
    return df_diff

# نمونه‌سازی دیتافریم‌های مربوط به جداول
data_current = {'Project': ['Task A', 'Task B'], 'Progress (%)': [80, 90]}
data_previous = {'Project': ['Task A', 'Task B'], 'Progress (%)': [75, 85]}

df_current = pd.DataFrame(data_current)
df_previous = pd.DataFrame(data_previous)

# مقایسه جداول
df_diff = compare_tables(df_current, df_previous)
print(df_diff)
```

```python
import os
from tkinter import Tk, Label, Button, filedialog, messagebox
from PIL import Image
import shutil

# تابع برای ایجاد پوشه ذخیره‌سازی عکس‌ها
def create_directory():
    if not os.path.exists("uploaded_images"):
        os.makedirs("uploaded_images")

# تابع برای انتخاب و ذخیره عکس‌ها
def upload_images():
    # باز کردن دیالوگ انتخاب فایل
    file_paths = filedialog.askopenfilenames(filetypes=[("Image Files", "*.jpg;*.jpeg;*.png;*.gif;*.bmp")])
    
    # بررسی اینکه کاربر عکسی را انتخاب کرده باشد
    if not file_paths:
        messagebox.showwarning("Warning", "No image selected!")
        return
    
    # ذخیره‌سازی عکس‌ها در پوشه
    for file_path in file_paths:
        try:
            # نام فایل و مسیر مقصد
            file_name = os.path.basename(file_path)
            destination = os.path.join("uploaded_images", file_name)
            
            # کپی فایل به پوشه مقصد
            shutil.copy(file_path, destination)
            
            # نمایش پیغام موفقیت
            messagebox.showinfo("Success", f"{file_name} saved successfully!")
        
        except Exception as e:
            messagebox.showerror("Error", f"An error occurred: {e}")

# ایجاد فرم اصلی
create_directory()
root = Tk()
root.title("Image Uploader")
root.geometry("400x200")

# برچسب عنوان
label = Label(root, text="آپلود تصاویر", font=("Helvetica", 16))
label.pack(pady=20)

# دکمه افزودن عکس
upload_button = Button(root, text="افزودن + ", command=upload_images, width=20, height=2)
upload_button.pack(pady=10)

# اجرای فرم
root.mainloop()
```

```python
import tkinter as tk
from tkinter import filedialog, messagebox
from tkinter import ttk
import cv2
from PIL import Image, ImageTk
import pytesseract
import matplotlib.pyplot as plt
import re

# تنظیم مسیر Tesseract (در صورت نیاز)
# pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

class App:
    def __init__(self, root):
        self.root = root
        self.root.title("تحلیل گزارش ماهیانه")
        self.root.geometry("800x600")

        # فریم ورودی
        self.frame_inputs = tk.Frame(self.root, pady=10)
        self.frame_inputs.pack()

        # ورودی عکس اول
        self.label_img1 = tk.Label(self.frame_inputs, text="عکس اول:")
        self.label_img1.grid(row=0, column=0, padx=5)
        self.entry_img1 = tk.Entry(self.frame_inputs, width=50)
        self.entry_img1.grid(row=0, column=1)
        self.btn_browse_img1 = tk.Button(self.frame_inputs, text="انتخاب فایل", command=self.load_img1)
        self.btn_browse_img1.grid(row=0, column=2, padx=5)

        # ورودی عکس دوم
        self.label_img2 = tk.Label(self.frame_inputs, text="عکس دوم:")
        self.label_img2.grid(row=1, column=0, padx=5)
        self.entry_img2 = tk.Entry(self.frame_inputs, width=50)
        self.entry_img2.grid(row=1, column=1)
        self.btn_browse_img2 = tk.Button(self.frame_inputs, text="انتخاب فایل", command=self.load_img2)
        self.btn_browse_img2.grid(row=1, column=2, padx=5)

        # ورودی موضوع
        self.label_topic = tk.Label(self.frame_inputs, text="موضوع:")
        self.label_topic.grid(row=2, column=0, padx=5)
        self.entry_topic = tk.Entry(self.frame_inputs, width=50)
        self.entry_topic.grid(row=2, column=1)

        # دکمه اجرای تحلیل
        self.btn_analyze = tk.Button(self.root, text="تحلیل و نمایش", command=self.analyze)
        self.btn_analyze.pack(pady=10)

        # نگهداری مسیر فایل‌ها
        self.img1_path = None
        self.img2_path = None

    def load_img1(self):
        self.img1_path = filedialog.askopenfilename(filetypes=[("Image files", "*.png;*.jpg;*.jpeg")])
        if self.img1_path:
            self.entry_img1.delete(0, tk.END)
            self.entry_img1.insert(0, self.img1_path)

    def load_img2(self):
        self.img2_path = filedialog.askopenfilename(filetypes=[("Image files", "*.png;*.jpg;*.jpeg")])
        if self.img2_path:
            self.entry_img2.delete(0, tk.END)
            self.entry_img2.insert(0, self.img2_path)

    def extract_text(self, image_path):
        # بارگذاری و پردازش تصویر
        image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
        image = cv2.medianBlur(image, 3)
        text = pytesseract.image_to_string(image, lang='fas')
        return text

    def analyze(self):
        if not self.img1_path or not self.img2_path:
            messagebox.showerror("خطا", "لطفاً هر دو تصویر را انتخاب کنید.")
            return

        # استخراج متن از هر دو تصویر
        text1 = self.extract_text(self.img1_path)
        text2 = self.extract_text(self.img2_path)

        # شناسایی اعداد و درصدها
        data1 = re.findall(r'\d+', text1)
        data2 = re.findall(r'\d+', text2)

        # مقایسه داده‌ها (برای نمایش ساده، اختلاف اعداد را محاسبه می‌کنیم)
        differences = [abs(int(d1) - int(d2)) for d1, d2 in zip(data1, data2)]

        # نمایش گرافیکی
        self.show_graph(data1, data2, differences)

    def show_graph(self, data1, data2, differences):
        plt.figure(figsize=(10, 5))

        # نمودار داده‌های تصویر اول و دوم
        plt.plot(range(len(data1)), [int(x) for x in data1], label="تصویر اول", marker='o')
        plt.plot(range(len(data2)), [int(x) for x in data2], label="تصویر دوم", marker='o')

        # نمودار اختلاف‌ها
        plt.bar(range(len(differences)), differences, alpha=0.5, label="اختلاف")

        plt.xlabel("شاخص‌ها")
        plt.ylabel("مقادیر")
        plt.title("مقایسه گزارش‌ها")
        plt.legend()
        plt.grid()
        plt.show()

# اجرای برنامه
root = tk.Tk()
app = App(root)
root.mainloop()
```
